## Program to fetch all the links present in the main URL "https://www.resetera.com/threads/media-create-sales-archive-thread.4115/#post-28337280"¶

```python
import requests
from bs4 import BeautifulSoup
url = "https://www.resetera.com/threads/media-create-sales-archive-thread.4115/#post-28337280"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
links = soup.find_all('a')
req_link=[]
for link in links:
    href = link.get('href')
    if href:
        req_link.append(href)
final_links=[]
for i in req_link:
    if 'www.resetera.com' in i:
        final_links.append(i)
```

## Function which scrapes and brings all the data into a list when given a link.

```python
def webscraper(url):
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text, 'lxml')
    lines=str()
    for points in soup.find_all('div',{"class":"bbWrapper"}):
        point = str(points.text)
        lines=lines+point
    a=lines.split(',')
    req_lines = lines
    final_list=(re.split('\n\d', req_lines))
    req_list=[]
    for i in final_list:
        for j in i:
            if j=='[':
                req_list.append(i)
    return req_list
```

## Loop to scrape every link present in the main link

```python
final_data=[]
for i in final_links:
    final_data.append(i)
    a=webscraper(i)
    final_data=final_data+a
```

## Loop to remove unwanted data lines from the final data

```python
pattern = r'\[[a-zA-Z0-9]{3}\]'
final_scraped_data=[]

# loop over lines and print lines that match the pattern
for line in added_lines:
    if re.search(pattern, line):
        final_scraped_data.append(line)
```
